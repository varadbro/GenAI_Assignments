```python
# Step 1: Install necessary libraries
!pip install transformers diffusers torch accelerate

# Step 2: Import libraries
import torch
from diffusers import StableDiffusionPipeline

# Step 3: Load the model from Hugging Face
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Move model to GPU

# Step 4: Define a function to generate images from text prompts
def generate_image(prompt):
    # Generate image
    image = pipe(prompt).images[0]
    return image

# Step 5: Input text prompt and generate an image
prompt = "A futuristic city at sunset, with flying cars and neon lights"
generated_image = generate_image(prompt)
```
