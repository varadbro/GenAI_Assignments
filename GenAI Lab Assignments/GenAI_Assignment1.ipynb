```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
# Generate synthetic data
np.random.seed(42)
data = np.vstack([
    np.random.multivariate_normal([2, 2], [[1, 0.5], [0.5, 1]], 180),
    np.random.multivariate_normal([-2, -2], [[1, -0.5], [-0.5, 1]], 120)
])
```

```python
# Fit Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(data)
labels = gmm.predict(data)
```

```python
# Visualize results
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=10, alpha=0.5)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='x', s=100)
plt.show()
```
