```python
from transformers import pipeline

# Load a Hugging Face model for question answering
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

def get_answer(question, context):
    """Function to get answer from the QA model"""
    response = qa_pipeline(question=question, context=context)
    return response["answer"]

# Example context for the chatbot
context = """
Climate change refers to long-term shifts in temperatures and weather patterns. These shifts may be natural, but since the 1800s, human activities have been the main driver, primarily due to burning fossil fuels like coal, oil, and gas.
"""

while True:
    question = input("Ask me a question about climate change (or type 'exit' to stop): ")
    if question.lower() == 'exit':
        break
    answer = get_answer(question, context)
    print("Answer:", answer)
```
